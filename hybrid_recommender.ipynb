## Collaborative Filtering

```python
x= new_books.groupby('User-ID').count()['Book-Rating'] > 10
padhe_likhe_users=x[x].index
```

```python
filtered_rating = new_books[new_books['User-ID'].isin(padhe_likhe_users)]
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=10
famous_books=y[y].index
famous_books
```

```python
final_ratings= filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt= final_ratings.pivot_table(index='Book-Title', columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores
```

```python
def collab_filt_recommend(book_name):
    #index fetch
    index = np.where(pt.index == book_name)[0][0]
    
    similar_items= sorted(list(enumerate(similarity_scores[index])),key = lambda x:x[1],reverse=True)[1:6]
    L=[]
    for i in similar_items:
        L.append(pt.index[i[0]])
    return L
```

```python
collab_list=collab_filt_recommend("The Kitchen God's Wife")
collab_list
```

## Content-Based Filtering

```python
books_content
```

```python
book = books_content[['Book','Author','Description','Genres']]

```

```python
book.isnull().sum()
```

```python
book.dropna(inplace=True)
```

```python
book.isnull().sum()
```

```python
book.head()
```

```python
book['Genres']
```

```python
#book['Genres']=book['Genres'].apply(convert)
book.head()
```

```python
book.loc[:, 'Description'] =book['Description'].apply(lambda x:x.split(" "))
```

```python
book.loc[:,'Description']=book['Description'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
book['Author']=book['Author'].apply(lambda x:x.replace(" ",""))

#book.loc[:,'Author'] = book['Author'].apply(lambda author: ''.join(author))
```

```python
book['Genres']=book['Genres'].apply(lambda x:x.replace(" ",""))
```

```python
book['Author'] = book['Author'].apply(lambda x: [x])
```

```python
book.head()
```

```python
def fix_case(genres):
    
    genre_list = eval(genres)
    
    return genre_list 
```

```python
book['Genres']=book['Genres'].apply(fix_case)
```

```python
book['tags']=book['Description']+book['Author']+book['Genres']
```

```python
new_df = book[['Book','tags']]
```

```python
new_df.head()
```

```python
new_df.loc[0, 'tags']
```

```python
new_df['tags']=new_df['tags'].apply(lambda x :" ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x :x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
len(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[7])),reverse = True, key=lambda x :x[1])[1:6]
```

```python
def content_based_recommend(book):
    
    book_index = new_df[new_df['Book'] == book].index[0]
    distances = similarity[book_index]
    book_list= sorted(list(enumerate(distances)),reverse = True, key=lambda x :x[1])[1:6]
    L=[]
    for i in book_list:
       
        L.append(new_df.iloc[i[0]].Book)
    return L
```

```python
c_books=content_based_recommend("The Kitchen God's Wife")
c_books
```

```python
book['Book']
```

```python
def hybrid_recommender(book):
    try:
      content_books = content_based_recommend(book) 
    except:
      content_books=[]
    
    try:
        collab_books = collab_filt_recommend(book)
    except:
       collab_books=[]
    return content_books+collab_books
    
    
```

```python
hybrid_books=hybrid_recommender("The Diary of a Young Girl")
```
